```python
#IMPORT REQUIRED LIBRARIES

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
from textblob import TextBlob

#LOAD THE DATASET
df = pd.read_csv('twitter_training.csv')
df.head()

#CREATE DATAFRAME
df = pd.DataFrame(df[['im getting on borderlands and i will murder you all ,','Positive']])
df.head()

# SENTIMENT ANALYSIS FUNCTION
def get_sentiment(text):
    text = str(text)
    analysis = TextBlob(text)
    if analysis.sentiment.polarity > 0:
        return 'Positive'
    elif analysis.sentiment.polarity == 0:
        return 'Neutral'
    else:
        return 'Negative'

# APPLYING SENTIMENT ANALYSIS
df['Sentiment'] = df['im getting on borderlands and i will murder you all ,'].apply(get_sentiment)
df.head()

# PRINT SENTIMENT ANALYSIS
print("\nSentiment Analysis Dataframe:")
print(df)

# VISUALIZING SENTIMENT ANALYSIS
plt.figure(figsize=(8, 5))
sns.countplot(x="Sentiment",hue="Sentiment",data=df, palette=["blue","green","red"],legend=False)
plt.title("Sentiment Distribution")
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.show()
```
